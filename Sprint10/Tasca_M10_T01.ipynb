#### Beautiful Soup

```python
# importem les llibreries
import requests
from bs4 import BeautifulSoup
```

```python
# Accedim a la pàgina web
URL = "http://quotes.toscrape.com"

page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
soup
```

La informació extreta encara està en codi html. Pel que ens interessa de la pàgina buscarem en les etiquetes <div> on hi hhagi la classe anomenada 'container'

```python
results = soup.find("div", class_="container")
results
```

Serem una mica més específics i buscarem dins els *results* la classe 'quote'

```python
quotes_elements = results.find_all("div", class_="quote")
quotes_elements
```

Ara ja podem veure les frases i els autors. També obtindrem les etiquetes amb les quals han estat categoritzades les frases:

```python
for quotes_element in quotes_elements:
    title_element = quotes_element.find(class_="text")
    by = quotes_element.find(class_="author")
    tags = quotes_element.find(class_="tags")
    print(title_element.text)
    print(by.text)
    print(tags.text)
    print('')
```

### Selenium

```python
# Importem les llibreríes necessàries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service as FirefoxService
from webdriver_manager.firefox import GeckoDriverManager

import pandas as pd

driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
```

Accedim a la url mitjançant el driver.get

```python
url = 'https://www.bolsasymercados.es/bme-exchange/es/Mercados-y-Cotizaciones/Acciones/Mercado-Continuo/Precios/ibex-35-ES0SI0000005'

driver.get(url)
```

Per tal d'accedir a la informació utilitzarem la nomenclatura xpath, en aquest cas

'//*[contains(@class,"shares-table")]//tr'

On estem dient que volem tots els elements (//\*) que continguin l'estiqueta @class igual a 'shares-table'per totes les generacions futures *tr* (//tr).

Finalment apendem les dades amb tots els elements (//\*) que siguin del tipus td o th (self::td or self::th])

```python
rows = []
for table in driver.find_elements(by= By.XPATH, value = '//*[contains(@class,"shares-table")]//tr'):
        data = [item.text for item in table.find_elements(by= By.XPATH, value= ".//*[self::td or self::th]")]
        # afegim les dades per files
        rows.append(data)
        
# la primera filera és la que pertany als noms de les columnes.
stock_indices = pd.DataFrame(rows[1:], columns = rows[0])
stock_indices
```

### Scrapy

Finalment farem *scrap* d'arribada de vols a l'aeroport de Barcelona utilitzant Scrapy. Amb l'Scrapy és comú codificar *spiders*. Aqui accedirem a la pàgina de l'aeroprt de Barcelona en poques líneas. Enlloc d'utilitzar la nomenclatura xpath com anteriorment farem servir la nomenclatura css, molt utilitzada també en web scrapping.

```python
#importem les llibreríes
from scrapy import Selector
from scrapy import Request
import requests
```

```python
#Accedim a la pàgina wev
html = requests.get( 'https://www.barcelona-airport.com/eng/arrivals.php' ).content
```

```python
# Creem un selector
sel = Selector( text=html )
```

Selleccionem les dades que incluirem en el df. En la nomenclatura css (.class) és el nom de la classe i amb ::text accedirem i extreurem el text. de forma similar al xpath les barres (/) són substituides per el símbol (>) i la doble barra (//) amb un espai.

```python
hour = sel.css(".flight-col__hour ::text").extract()
city = sel.css(".flight-col__dest-term b:nth-child(1) ::text").extract()
num = sel.css(".flight-col__flight > a:nth-child(1) ::text").extract()
airline = sel.css(".flight-col__airline > a:nth-child(1) ::text").extract()
terminal = sel.css(".flight-col__terminal-mob ::text").extract()
status = sel.css(".flight-col__status > a ::text").extract()

hour = list(map(lambda x:x.strip(),hour))
```

```python
import pandas as pd
arrivals_bcn = pd.DataFrame(
    {'hour': hour.pop(0)[0:100],
    'city': city[0:100],
     'num': num[0:100],
     'airline':airline[0:100],
     'terminal':terminal[0:100],
     'status':status[0:100]
    })
```
